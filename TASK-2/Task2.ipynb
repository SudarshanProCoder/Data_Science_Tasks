```python
import pandas as pd

# Load the dataset
file_path = 'test.csv'
data = pd.read_csv(file_path)


# Display the first few rows of the dataset to understand its structure
print(data.head())
print(data.info())
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())

# Drop rows with missing values or fill them with appropriate values
data_cleaned = data.dropna()
# Check for duplicates
print(data_cleaned.duplicated().sum())

# Remove duplicates if any
data_cleaned = data_cleaned.drop_duplicates()

# Ensure data types are appropriate
print(data_cleaned.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of a numerical variable
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between Age and Survived

plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=data_cleaned)
plt.title('Age vs. Survived')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

# Relationship between Sex and Survived
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Sex', data=data_cleaned)
plt.title('Survived vs. Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Pairplot to explore relationships between multiple variables
sns.pairplot(data_cleaned, hue='Survived')
plt.show()

# Correlation matrix to understand relationships between numerical variables
# plt.figure(figsize=(12, 8))
# sns.heatmap(data_cleaned.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
# plt.title('Correlation Matrix')
# plt.show()

# plt.figure(figsize=(12, 8))
# sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
# plt.title('Correlation Matrix')
# plt.show()
```
