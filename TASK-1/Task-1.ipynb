```python
import pandas as pd

# Load the Excel file
file_path = 'P_Data_Extract_From_World_Development_Indicators.xlsx'
xls = pd.ExcelFile(file_path)

# Check the sheet names to understand the structure of the file
sheet_names = xls.sheet_names
print(sheet_names)

# Load the 'Data' sheet into a DataFrame
data_df = pd.read_excel(xls, sheet_name='Data')

# Display the first few rows and the columns of the DataFrame to understand its structure
print(data_df.head())
print(data_df.columns.tolist())
```

```python
import matplotlib.pyplot as plt

# Extract the population data for 2023 and convert it to numeric
population_2023 = pd.to_numeric(data_df['2023 [YR2023]'], errors='coerce')

# Drop NaN values
population_2023_cleaned = population_2023.dropna()

# Create a histogram of the population distribution for 2023
plt.figure(figsize=(10, 6))
plt.hist(population_2023_cleaned, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Population Across Countries in 2023')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Hypothetical data: Gender distribution in a population
genders = ['Male', 'Female']
counts = [480, 520]

# Create a bar chart for the gender distribution
plt.figure(figsize=(8, 5))
plt.bar(genders, counts, color=['blue', 'pink'])
plt.title('Distribution of Genders in a Population')
plt.xlabel('Gender')
plt.ylabel('Number of People')

# Show the plot
plt.show()
```
